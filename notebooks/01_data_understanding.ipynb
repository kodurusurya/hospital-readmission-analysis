## 6. Data Quality Assessment

### Missing Value Assessment

Initial inspection using `isnull()` suggested that the dataset contained very few missing values. However, further investigation revealed that several categorical variables use the placeholder `"?"` instead of standard null values.

The most affected columns are:

- weight (~97% missing)
- medical_specialty (~49% missing)
- payer_code (~40% missing)

These variables will be evaluated during the data cleaning phase to determine whether they should be retained, imputed, or removed.

## 7. Summary of Findings

Key Findings

The dataset contains 101,766 hospital encounters across 50 variables.
Each row represents a unique hospital encounter; no duplicate encounter records were found.
Approximately 30,248 patients have multiple encounters, indicating repeat hospital visits.
Several variables use "?" as a placeholder for missing values instead of standard null values.
weight, medical_specialty, and payer_code have substantial missingness and will require evaluation during the cleaning phase.
The target variable (readmitted) is imbalanced, with only about 11% of encounters resulting in readmission within 30 days.
